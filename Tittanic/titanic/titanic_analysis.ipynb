```python
def encode_cabins(cabin):
    if pd.isna(cabin):
        return 0  # Pas de cabine (NaN)
    elif cabin[0] in ['B', 'C']:
        return 3  # Cabines commençant par B ou C
    else:
        return 1  # Autres cabines

def encode_cabins_2(cabin):
    if pd.isna(cabin):
        return 0  # Pas de cabine (NaN)
    else:
        return cabin[0]  # Autres cabines
```

```python
def extract_title(name):
    title = name.split(',')[1].split('.')[0].strip()
    return title

tdfx = train_data.copy()
# Appliquez la fonction pour extraire les titres
tdfx['Title'] = tdfx['Name'].apply(extract_title)

# Affichez les titres uniques pour voir quels titres ont été extraits
print(tdfx['Title'].unique())
tdfx['Title'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
def prepare_data_titanic(df):
    df_processed = df.copy()
    # Drop columns which will be unused
    df_processed = df_processed.drop(['PassengerId', 'Name', 'Ticket'] , axis=1)

    # Encodage des variables catégorielles
    df_processed = pd.get_dummies(df_processed, columns=["Sex", "Embarked"])
    df_processed['Cabin'] = df_processed['Cabin'].apply(encode_cabins)

    # Remplissage des variables null (age par la moyenne)
    df_processed = df_processed.fillna(df_processed['Age'].mean())

    # Mise à l'échelle des données
    scaler = MinMaxScaler()
    df_processed = pd.DataFrame(scaler.fit_transform(df_processed), columns=df_processed.columns)

    return df_processed

def prepare_data_titanic_2_bis(df):
    df_processed = df.copy()

    # Ajout de variables catégorielles
    df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch']
    df_processed['IsAlone'] = 0
    df_processed.loc[df_processed['FamilySize'] == 0, 'IsAlone'] = 1
    df_processed['Title'] = df_processed['Name'].apply(extract_title)

    # Drop columns which will be unused
    df_processed = df_processed.drop(['PassengerId', 'Name', 'Ticket', "Embarked"] , axis=1)

    # Encodage des variables catégorielles
    df_processed = pd.get_dummies(df_processed, columns=["Sex", "Title"])
    df_processed['Cabin'] = df_processed['Cabin'].apply(encode_cabins)

    # Remplissage des variables null (age par la moyenne)
    df_processed = df_processed.fillna(df_processed['Age'].mean())


    
    # Mise à l'échelle des données
    scaler = MinMaxScaler()
    df_processed = pd.DataFrame(scaler.fit_transform(df_processed), columns=df_processed.columns)

    return df_processed

def prepare_data_titanic_2_ter(df):
    df_processed = df.copy()

    # Ajout de variables catégorielles
    df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch']
    
    df_processed['IsAlone'] = 0
    df_processed.loc[df_processed['FamilySize'] == 0, 'IsAlone'] = 1
    
    df_processed['FarePerPerson'] = df_processed['Fare'] / (df_processed['FamilySize'] + 1)
    
    age_bins = [0, 12, 18, 60, df['Age'].max()]  # Vous pouvez ajuster les plages d'âge au besoin
    age_labels = ['Enfants', 'Adolescents', 'Adultes', 'Personnes âgées']
    df_processed['AgeCategory'] = pd.cut(df_processed['Age'], bins=age_bins, labels=age_labels)

    
    # Drop columns which will be unused
    df_processed = df_processed.drop(['PassengerId', 'Name', 'Ticket', "Embarked"] , axis=1)

    # Encodage des variables catégorielles
    df_processed['Cabin'] = df_processed['Cabin'].apply(encode_cabins_2)
    df_processed = pd.get_dummies(df_processed, columns=["Sex", "Cabin", "AgeCategory"])

    # Remplissage des variables null (age par la moyenne)
    df_processed = df_processed.fillna(df_processed['Age'].mean())


    
    # Mise à l'échelle des données
    scaler = MinMaxScaler()
    df_processed = pd.DataFrame(scaler.fit_transform(df_processed), columns=df_processed.columns)

    return df_processed

def prepare_data_titanic_2(df, 
                           categorial_vars = [],
                           unused_var=['PassengerId', 'Name', 'Ticket', "Embarked", 'Cabin_T']):
    df_processed = df.copy()

    # Ajout de variables catégorielles
    df_processed['FamilySize'] = df_processed['SibSp'] + df_processed['Parch']
    df_processed['IsAlone'] = 0
    df_processed.loc[df_processed['FamilySize'] == 0, 'IsAlone'] = 1
    df_processed['FarePerPerson'] = df_processed['Fare'] / (df_processed['FamilySize'] + 1)
    df_processed['AgeMissing'] = df_processed['Age'].isnull().astype(int)
    df_processed['CabinMissing'] = df_processed['Cabin'].isnull().astype(int)
    


    # Encodage des variables catégorielles
    df_processed['Cabin'] = df_processed['Cabin'].apply(encode_cabins_2)
    df_processed = pd.get_dummies(df_processed, columns=["Sex", "Cabin"])

    # Remplissage des variables null (age par la moyenne)
    df_processed = df_processed.fillna(df_processed['Age'].mean())

    # Mise à l'échelle des données
    scaler = MinMaxScaler()

    categorial_vars_all = ["FamilySize", "IsAlone", "FarePerPerson", "AgeMissing", "CabinMissing"]
    for categorial_var in categorial_vars_all:
        if categorial_var not in categorial_vars:
            df_processed = df_processed.drop(categorial_var, axis=1)

    # Drop columns which will be unused
    columns_to_drop_existing = [col for col in unused_var if col in df_processed.columns]
    df_processed = df_processed.drop(columns_to_drop_existing , axis=1)
            
    df_processed = pd.DataFrame(scaler.fit_transform(df_processed), columns=df_processed.columns)

    return df_processed
train_data
```
